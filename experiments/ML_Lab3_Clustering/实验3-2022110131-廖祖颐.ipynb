```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Set the number of threads to avoid memory leakage
os.environ["OMP_NUM_THREADS"] = "2"

# Paths to dataset files
datasets = {
    "noisy_circles": "noisy_circles.txt",
    "noisy_moons": "noisy_moons.txt",
    "blobs": "blobs.txt",
    "aniso": "aniso.txt",
    "no_structure": "no_structure.txt"
}

def load_data(file_name):
    return np.loadtxt(file_name)

# Process each dataset
for dataset_name, file_name in datasets.items():
    # Load the data
    X = load_data(file_name)
    inertia = []
    
    # Perform K-means clustering and calculate inertia for different values of k
    for k in range(1, 8):
        kmeans = KMeans(n_clusters=k, n_init='auto', random_state=1).fit(X)
        inertia.append(np.sqrt(kmeans.inertia_))
    
    # Plot the inertia graph
    plt.figure()
    plt.plot(range(1, 8), inertia, marker='s')
    plt.xlabel('$k$')  # Number of clusters
    plt.ylabel('$J(C_k)$')  # Inertia
    plt.title(f'Elbow Method for Optimal k: {dataset_name}')
    plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# List of datasets
data_files = {
    "noisy_circles": "noisy_circles.txt",
    "noisy_moons": "noisy_moons.txt",
    "blobs": "blobs.txt",
    "aniso": "aniso.txt",
    "no_structure": "no_structure.txt"
}

# Set the value of k, typically equal to min_samples for DBSCAN
k = 5

# Process each dataset
for name, path in data_files.items():
    # Load data
    data = np.loadtxt(path)
    
    # Initialize and fit NearestNeighbors
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(data)
    
    # Retrieve distances to k-nearest neighbors for each point
    distances, indices = neigh.kneighbors(data)
    
    # Sort distances
    sorted_distances = np.sort(distances[:, k-1])
    
    # Compute the first derivative of the sorted distances
    first_derivatives = np.diff(sorted_distances, n=1)
    
    # Compute the second derivative from the first derivatives
    second_derivatives = np.diff(first_derivatives, n=1)
    
    # Find the index of the elbow point
    elbow_index = np.argmax(second_derivatives) + 1  # +1 to compensate for the diff reduction
    
    # Plot the k-distance graph
    plt.figure(figsize=(10, 5))
    plt.plot(sorted_distances, label='K-distance')
    plt.plot(elbow_index, sorted_distances[elbow_index], 'ro', label=f'Elbow Point at {elbow_index}')
    plt.title(f'K-distance Graph for {name}')
    plt.xlabel('Points sorted by distance')
    plt.ylabel(f'Distance to {k}-th nearest neighbor')
    plt.legend()
    plt.grid(True)
    plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
import pandas as pd

# Datasets file names and number of clusters
datasets = {
    "noisy_circles.txt": 2,
    "noisy_moons.txt": 2,
    "blobs.txt": 3,
    "aniso.txt": 3,
    "no_structure.txt": 3
}

# Set the eps parameters for DBSCAN
eps_values = {
    "noisy_circles.txt": 0.14,
    "noisy_moons.txt": 0.12,
    "blobs.txt": 1.75,
    "aniso.txt": 0.4,
    "no_structure.txt": 0.11
}

# Minimum number of samples per cluster for DBSCAN
min_samples = {
    "noisy_circles.txt": 4,
    "noisy_moons.txt": 4,
    "blobs.txt": 3,
    "aniso.txt": 4,
    "no_structure.txt": 5
}

# Load data and apply clustering algorithms
for file_name, n_clusters in datasets.items():
    data = np.loadtxt(file_name)
    
    # K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, n_init='auto')
    kmeans.fit(data)
    kmeans_labels = kmeans.labels_
    
    # DBSCAN clustering
    dbscan = DBSCAN(eps=eps_values[file_name], min_samples=min_samples[file_name])
    dbscan.fit(data)
    dbscan_labels = dbscan.labels_
    
    # Plot the original data
    plt.figure(figsize=(12, 6))
    plt.subplot(131)
    plt.scatter(data[:, 0], data[:, 1], s=10)
    plt.title(f'Original Data: {file_name}')
    
    # Plot results of K-means clustering
    plt.subplot(132)
    plt.scatter(data[:, 0], data[:, 1], c=kmeans_labels, s=10, cmap='viridis')
    plt.title('K-means Clustering')
    
    # Plot results of DBSCAN clustering
    plt.subplot(133)
    plt.scatter(data[:, 0], data[:, 1], c=dbscan_labels, s=10, cmap='viridis')
    plt.title('DBSCAN Clustering')
    
    plt.tight_layout()
    plt.show()
```
